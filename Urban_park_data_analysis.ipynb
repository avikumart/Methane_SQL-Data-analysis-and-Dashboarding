### How much time it took to respond to a call?

```python
df["Date and Time of initial call"] = pd.to_datetime(df["Date and Time of initial call"], format="%d%m%yyyy %H%M%S")
df["Date and time of Ranger response"] = pd.to_datetime(df["Date and time of Ranger response"], format="%d%m%yyyy %H%M%S")
```

```python
df["Duration of Response"] = df["Date and time of Ranger response"] - df["Date and Time of initial call"]
```

```python
df.info()
```

```python
df["Duration of Response"].plot()
```

```python
df["Duration of Response"].describe()
```

```python
df["Duration of Response in secs"] = df["Duration of Response"].dt.total_seconds()
```

```python
df["Duration of Response in secs"].plot(kind="hist",bins=50)
```
